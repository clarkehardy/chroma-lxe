```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('../macros/photonbomb_test.csv')
df.sort_values('n', inplace=True)

plt.plot(df['n'], df['pte'])
plt.xlabel('N')
plt.ylabel('PTE')
plt.title('Photon Bomb Test')
plt.xscale('log')
plt.yscale('log')
plt.twinx()
plt.plot(df['n'], df['time_spent'], color='red')
plt.ylabel('Time taken (s)')
plt.yscale('log')
plt.show()
```

We'll use 1e5 photons per position.

```python
import numpy as np
print('10^5 photons will take approx', np.interp(10**5, df['n'], df['time_spent']), 'sec')
```
