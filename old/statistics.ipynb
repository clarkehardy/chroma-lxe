```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('../data/photonbomb_test.csv')
df.sort_values('n', inplace=True)

plt.figure(dpi=300)
plt.plot(df['n'], df['pte'], color='blue')
plt.xlabel('N')
plt.ylabel('PTE')
plt.xscale('log')
plt.yscale('log')
plt.twinx()
plt.plot(df['n'], df['time_spent'], color='red')

# plot legend
lines = [plt.Line2D([0], [0], color='blue', label='PTE'),
            plt.Line2D([0], [0], color='red', label='Time spent')]
plt.legend(lines, [l.get_label() for l in lines],frameon=False)


plt.ylabel('Time taken (s)')
plt.yscale('log')
plt.show()
```
